```
#Create a mapping between id and title
idx2title = {int(row["Movie_Id"]): row["movie_title"] for _, row in movies.iterrows()}
title2idx = {j: i for i, j in idx2title.items()}
```

```
highRatings = ratings[ratings.Rating >= 4]
itemsets = [
    [idx2title[mov] for mov in highRatings[highRatings.User_Id == user].Movie_Id]
    for user in highRatings.User_Id
]
```

```
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
te_ary = te.fit(itemsets).transform(itemsets)

df = pd.DataFrame(te_ary, columns=te.columns_)
```

```
df.head()
```

```
from pymining import itemmining, assocrules, perftesting

# isets=[]
# df= isets
# assocrules.mine_assoc_rules

sup = 0.03 

relim_input = itemmining.get_relim_input(df)
# report = itemmining.relim(relim_input, min_support=2)
item_sets = itemmining.relim(relim_input, int(len(df) * sup))
rules = assocrules.mine_assoc_rules(item_sets, min_support=0.03, min_confidence=0.2)




```

```
rules
```

```
from mlxtend.frequent_patterns import apriori, association_rules

frequent_itemsets = apriori(df, min_support=0.03, use_colnames=True, max_len=2)

rules = association_rules(frequent_itemsets, metric="lift", min_threshold=2)
```
